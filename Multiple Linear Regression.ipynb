#### Problem Statement:

Consider that a real estate company has the data of real estate prices in Delhi. The company wants to optimise the selling price of the properties, based on important factors such as area, bedrooms, parking, etc.

 
> Essentially, the company wants: 

1. To identify the variables affecting house prices, e.g., area, number of rooms, bathrooms, etc.
2. To create a linear model that quantitatively relates house prices with variables, such as the number of rooms, area, number of bathrooms, etc.
3. To know the accuracy of the model, i.e. how well do these variables predict the house prices

##### Converting the categorical variables to binary and creating dummy variables

```python
# Identifying yes/no variable

var_cols = ['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
df[var_cols] = df[var_cols].apply(lambda x: x.map({'yes': 1, 'no':0}))
```

```python
df.head()
```

Creating dummy variables

```python
status = pd.get_dummies(df['furnishingstatus'], dtype =int)
status.head()
```

Here, we can simply understand as:
1. 00 = Furnished
2. 01 = Unfurnished
3. 10 = Semi-Furnished

Dropping reduntant furnishing status

```python
status.drop(columns = 'furnished', inplace=True)
```

```python
status.head()
```

Joining the status with df

```python
df = pd.concat([df, status], axis= 1)
df.head()
```

dropping the original column

```python
df.drop(columns = 'furnishingstatus', inplace = True)
df.head()
```

##### Train Test Split

```python
df_train, df_test  = train_test_split(df, train_size = 0.7, random_state = 100)
print(df_train.shape)
print(df_test.shape)
```

Rescaling the variable
1. Min-Max Scaling (normalisation) : Compress the values b/w 0 and 1
2. Standardisation

```python
scaler  = MinMaxScaler()

num_vars = ['area','bedrooms','bathrooms','stories','parking','price']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
df_train.head()
```

##### Building and Training the model

```python
# Heatmap
plt.figure(figsize=(12,12))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

Creating X_train and y_train

```python
X_train = df_train
y_train = df_train.pop('price')
```

```python
X_train_sm = sm.add_constant(X_train['area'])
lr = sm.OLS(y_train, X_train_sm)
lr_model = lr.fit()
lr_model.params
```

```python
lr_model.summary()
```

```python
# Adding another variable
X_train_sm = X_train[['area','bathrooms']]
X_train_sm = sm.add_constant(X_train_sm)

lr = sm.OLS(y_train, X_train_sm)
lr_model = lr.fit()
lr_model.params
```

```python
lr_model.summary()
```

P value of all the three ae zero which means, it is statistically significant

```python
# Adding other variable
X_train_sm = X_train[['area','bathrooms','bedrooms']]
X_train_sm = sm.add_constant(X_train_sm)

lr = sm.OLS(y_train, X_train_sm)
lr_model = lr.fit()
lr_model.summary()
```

```python
# Adding other variable
X_train_sm = X_train[['area','bathrooms','bedrooms', 'airconditioning']]
X_train_sm = sm.add_constant(X_train_sm)

lr = sm.OLS(y_train, X_train_sm)
lr_model = lr.fit()
lr_model.summary()
```

Baed on the p-value, we have to drop a variable along with by checking VIF

#### VIF

```python
vif = pd.DataFrame()
vif['Features'] = X_train.columns
vif['VIF'] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by='VIF', ascending =False)
vif
```

Variable drop selection based on

- High p-value, high VIF = Drop
- High-low:
  - low p, High VIF:
    - Drop these in the second place
  - high p, low VIF:
    - Drop these variables in the first place
- low p-low vif = Drop

```python
X = X_train.drop('semi-furnished', axis=1)

X_train_sm = sm.add_constant(X)

lr = sm.OLS(y_train, X_train_sm)
lr_model = lr.fit()
lr_model.summary()
```

```python
# Checking VIF
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by='VIF', ascending =False)
vif
```

```python
X = X_train.drop('bedrooms', axis=1)

X_train_sm = sm.add_constant(X)

lr = sm.OLS(y_train, X_train_sm)
lr_model = lr.fit()
lr_model.summary()
```

```python
# Checking VIF
vif = pd.DataFrame()
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'],2)
vif = vif.sort_values(by='VIF', ascending =False)
vif
```

##### Residual Analysis

```python
res = y_train-y_train_pred
sns.distplot(res)
```
